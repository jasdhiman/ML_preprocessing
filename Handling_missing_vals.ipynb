## Handling missing values

### DELETING ROWS COLUMNS _ MISSING VALS

```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv(r"C:\DS\house\train.csv")
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(df.isnull())
```

```python
# getting percentage of missing data per column
null_var = (df.isnull().sum()/df.shape[0])*100
null_var
```

```python
# getting > 17% missing values
drop_columns = null_var[null_var > 17].keys()
drop_columns
```

```python
#dropping columns
df2_drop_clm = df.drop(columns = drop_columns)
df2_drop_clm.shape
```

```python
sns.heatmap(df2_drop_clm.isnull())
```

```python
# dropping rows
df3_drop_rows = df2_drop_clm.dropna()
df3_drop_rows.shape
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(df3_drop_rows.isnull())
```

```python
# total missing values (double sum)
df3_drop_rows.isnull().sum().sum()
```

```python
# getting column names for numeric variables
df3_drop_rows.select_dtypes(include = ['int64', 'float64']).columns
```

```python
# distribution plot for numeric features - for cleaned and unclean dfs
sns.distplot(df['MSSubClass'])
sns.distplot(df3_drop_rows['MSSubClass'])
```

```python
# getting list of numeric variables
numvar_list = df3_drop_rows.select_dtypes(include = ['int64', 'float64']).columns
numvar_list
```

```python
numvar_list = numvar_list.drop('Id')
```

```python
numvar_list = list(numvar_list)
```

```python
plt.figure(figsize = (25, 25))
for i, var in enumerate(numvar_list):
    plt.subplot(9,4,i+1)
    sns.distplot(df[var], bins = 20)
    sns.distplot(df3_drop_rows[var], bins = 20)
```

```python
# getting cat vars
df3_drop_rows.select_dtypes(include = ['object']).columns
```

```python
#getting side by side
pd.concat([(df['MSZoning'].value_counts()/df.shape[0])*100, 
          (df3_drop_rows['MSZoning'].value_counts()/df3_drop_rows.shape[0])*100], 
          axis=1, 
          keys = ['MSZoning_orig', 'MSZoning_clean'])
```

```python
# defining function
def catvardist(var):
    return pd.concat([(df[var].value_counts()/df.shape[0])*100, 
          (df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0])*100], 
          axis=1, 
          keys = [var+'_orig', var+'_clean'])
```

```python
catvardist('MSZoning')
```

```python
# going through all cat vars
catlist = list(df3_drop_rows.select_dtypes(include = ['object']).columns)
for var in catlist:
    print(catvardist(var))
```
