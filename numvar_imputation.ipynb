```python
# now looping through different numerical variables imputed with mean values based on different categorical variables
# so we have 3 missing values containing numerical variables - missvars
# check
missvars
```

```python
# deciding what categorical variables these will be related to
df.select_dtypes(include = ['object']).keys()
```

```python
# so LotFrontage is associated with LotConfig (assumptions)
# MasVnrArea is related to MasVnrType
# GarageYrBlt is related to Garagetype
# numeric list is missvars
numlist = missvars.copy()
catlist = ['LotConfig', 'MasVnrType', 'GarageType']
```

```python
# using for loop with update option
df_copy_mean = df.copy()
for var_cat, var_num in zip(catlist, numlist):
    for var_unique in df[var_cat].unique():
        mean_val = df[df.loc[:, var_cat] == var_unique][var_num].mean()
        df_copy_mean.update(df[df.loc[:, var_cat] == var_unique][var_num].replace(np.nan, mean_val))
```

```python
# checking missing values now
df_copy_mean[numlist].isnull().sum()
```

```python
# we still have missing values in MasVnrArea and the 3rd variable
# let us check why
df_copy_mean[df_copy_mean['MasVnrArea'].isnull()]
```

```python
# we see that the values were not imputed because the categorical feature it was based on is also null at these locations
# checking for the other feature
df_copy_mean[df_copy_mean['GarageYrBlt'].isnull()]
```

```python
# same thing here - GarageType is all missing values at these locations
# LESSON - categorical feature cannot clean numeric features in relation to imputation if it itself is not clean
# trying other cat features (randomly selected - the ones which have no missing values)
df.select_dtypes(include = ['object']).isnull().sum()
```

```python
# let us use paveddrive for garage and bldgtype for MasVnr
# using for loop with update option
numlist = missvars.copy()
catlist = ['LotConfig', 'BldgType', 'PavedDrive']
df_copy_mean = df.copy()
for var_cat, var_num in zip(catlist, numlist):
    for var_unique in df[var_cat].unique():
        mean_val = df[df.loc[:, var_cat] == var_unique][var_num].mean()
        df_copy_mean.update(df[df.loc[:, var_cat] == var_unique][var_num].replace(np.nan, mean_val))
```

```python
# checking
df_copy_mean[numlist].isnull().sum()
```

```python
# clear now
# doing same for median
numlist = missvars.copy()
catlist = ['LotConfig', 'BldgType', 'PavedDrive']
df_copy_median = df.copy()
for var_cat, var_num in zip(catlist, numlist):
    for var_unique in df[var_cat].unique():
        median_val = df[df.loc[:, var_cat] == var_unique][var_num].median()
        df_copy_median.update(df[df.loc[:, var_cat] == var_unique][var_num].replace(np.nan, median_val))
```

```python
# checking
df_copy_median[numlist].isnull().sum()
```

```python
# checking distribution of 3 datasets: original, mean imputed, and median imputed
plt.figure(figsize=(15,10))
for i, var in enumerate(numlist):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], label='orig')
    sns.distplot(df_copy_mean[var], label='mean')
    sns.distplot(df_copy_median[var], label='median')
    plt.legend()
```

### The imputations did not change the distribution of the numeric variables from the original significantly. This shows that imputations are OK

```python
# checking if imputations changed/added/deleted outliers
for var in numlist:
    plt.figure()
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df_copy_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df_copy_median[var])
```

### The imputations did not change, add, or delete outliers from the original significantly. This also shows that imputations are OK
